# Ensemble

```python
import pandas as pd
import numpy as np
```

## Summed Ensemble

```python
submissions = [
    "submission1",
    "submission2",
]
csvs = {
    i : pd.read_csv(f"{i}.csv").sort_values("Sample_ID").reset_index(drop=True)
    for i in submissions
}
sub = csvs[submissions[0]][['Sample_ID']]
for s in submissions:
  sub[s] = csvs[s].Label 

sub
```

```python
sub['Label'] = (sub.eval(
      " + ".join(submissions)
  ) >= 1)*1

sub = sub[['Sample_ID', 'Label']]
print(sub.Label.value_counts())
sub
```

```python
sub.to_csv("ensemble.csv", index=False)
```
